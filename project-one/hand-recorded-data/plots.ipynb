```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from lmfit import Model, Parameters
import matplotlib.font_manager
from matplotlib.pyplot import figure
plt.rcParams.update({'font.size': 18})
plt.rcParams.update({
    "text.usetex": True,
    "font.weight": "bold",
    "font.family": "serif",
    "font.sans-serif": ["Helvetica"]})
```

```python
#Read the data

sample    = pd.read_csv("Readings/data.csv")
```

```python
#Make the loaded data a numpy array to do easy computations

sample = np.array(sample)
```

```python
#Read each column of the array differently

sample_freq       = sample[:,0]
sample_amp_1      = sample[:,1]
sample_amp_2      = sample[:,2]
sample_phase_diff = sample[:,3]
sample_amp_diff   = sample[:,4]
```

```python
sample_phase_diff.min(), sample_phase_diff.max()
```

```python
#Plot the Phase Difference Against Frequency 

plt.plot(sample_freq,sample_phase_diff, '.' , lw=0.5, c='chocolate')
plt.axvline(x=2.1, ymin=58.1, ymax=0, ls='dotted')
plt.grid()
plt.title("Phase Shift vs Frequency")
plt.xlabel("Frequency (KHz)")
plt.ylabel("Phase Shift (degrees)")
plt.savefig("Images/phaseshift-f.pdf", bbox_inches='tight')
plt.show()
```

```python
#Plot the Amplitude Difference Against Frequency 

plt.plot(sample_freq,sample_amp_diff, '--' ,lw=1, color='red')
plt.axvline(x= 17, ymin=0, ymax=1.8, ls='dotted')
plt.grid()
plt.title("Amplitude Difference vs Frequency")
plt.xlabel("Frequency (KHz)")
plt.ylabel("Amplitude Difference (V)")
plt.savefig("Images/amp-f.pdf", bbox_inches='tight')
plt.show()
```
