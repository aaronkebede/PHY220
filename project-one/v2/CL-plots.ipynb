```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager
import pandas as pd
plt.rcParams.update({'font.size': 16})
plt.rcParams.update({
    "text.usetex": True,
    "font.weight": "bold",
    "font.family": "serif",
    "font.sans-serif": ["Helvetica"]})
```

```python
#Import Data

CL_data = pd.read_csv("CircuitLabData/data.csv")
CL_data_2 = pd.read_csv("CircuitLabData/data-2.csv")
R = 1.5e6 #Resistance of the circuit was 1.5 M Ohms
C = 22e-9 # The capacitance was 0.022 microFarads
```

```python
# Load data into array 

CL_Array = np.array(CL_data)
CL_Array_2 = np.array(CL_data_2)
```

```python
#Specify data

Time    = CL_Array[:,0]
Time_2  = CL_Array_2[:,0]
Voltage = CL_Array[:,1]
Voltage_2 = CL_Array_2[:,1]
```

```python
tau = R*C
```

```python
V_tau =(1-(1/np.e))*Voltage.max()
```

```python
plt.plot(Time, Voltage, '--', lw=1.2, color='tomato')
plt.xlabel("Time(sec)")
plt.ylabel("Voltage(V)")
plt.axvline(tau, lw='0.6', ls='-', c="green", label="tau")
plt.axhline(V_tau, lw='0.6', c="blue", label="V(tau)")
plt.legend()
plt.savefig("Images/RC Circuit Plot.pdf", bbox_inches='tight')
plt.show()
```

```python
plt.plot(Time_2, Voltage_2, '-', lw=2, color='m')
plt.xlabel("Time(sec)")
plt.ylabel("Voltage(V)")
plt.axvline(tau, lw='0.6', ls='-', c="green", label="tau")
plt.axhline(V_tau, lw='0.6', c="blue", label="V(tau)")
plt.legend()
plt.savefig("Images/Signal Generated RC Circuit Plot.pdf", bbox_inches='tight')
plt.show()
```
